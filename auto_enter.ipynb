```python
import pyautogui
import time
import sys

# 이 스크립트를 실행한 후, 작업을 시작하기 전에 5초의 시간을 줍니다.
# 이 시간 동안 마우스를 화면 구석(보통 왼쪽 상단)으로 옮기면 프로그램이 비상 정지(Fail-safe)됩니다.
# (Fail-safe 기능을 사용하지 않으려면 pyautogui.FAILSAFE = False 로 설정할 수 있습니다.)
print("5초 후에 프로그램이 시작됩니다. 이 시간 동안 다른 창을 활성화하여 엔터 키가 눌릴 위치를 지정하세요.")
time.sleep(5)
print("프로그램 시작!")

try:
    while True:

        
        # 엔터 키를 누르는 명령
        pyautogui.press('enter')
        print(f"[{time.strftime('%H:%M:%S')}] 엔터 키를 눌렀습니다.")
        
        # 180초 대기
        time.sleep(300)


except KeyboardInterrupt:
    # 사용자가 Ctrl+C를 눌러 수동으로 종료할 때
    print("\n[알림] 사용자의 요청으로 프로그램을 종료합니다.")
except pyautogui.FailSafeException:
    # 마우스를 화면 모서리로 옮겨 프로그램이 비상 정지되었을 때
    print("\n[경고] 마우스 커서가 화면 모서리에 닿아 프로그램이 비상 정지(Fail-safe)되었습니다. 종료합니다.")
except Exception as e:
    print(f"\n[오류] 예기치 않은 오류가 발생했습니다: {e}")
finally:
    sys.exit()
```
